### Sentiment Analysis

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("I loved Star Wars so much!")
```

### Grammatical Correctness

```python
from transformers import pipeline

classifier = pipeline("text-classification", model = "textattack/distilbert-base-uncased-CoLA")
classifier("I will walk to home when I went through the bus.")
```

In the CoLA model:

    LABEL_1 = unacceptable

    LABEL_0 = acceptable

### Text Generation and Corrector

```python
from transformers import pipeline

corrector = pipeline("text2text-generation", model="vennify/t5-base-grammar-correction")

corrector("I will walk to home when I went through the bus.")
```

### Question Answering

```python
from transformers import pipeline

qa_model = pipeline("question-answering")
question = "Where do I live?"
context = "My name is Merve and I live in İstanbul."
qa_model(question = question, context = context)
```

### Translation

```python
from transformers import pipeline

en_es_translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-es")
en_es_translator("Hello, How are you?")
```

```python
es_en_translator = pipeline("translation", model="Helsinki-NLP/opus-mt-es-en")
es_en_translator("Hola, ¿cómo estás?")
```
